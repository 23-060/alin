# Transformasi Linear 
Transformasi linear adalah pemetaan matematika yang mempertahankan sifat garis lurus dan operasi linear. Dalam konteks ruang vektor, transformasi linear menghubungkan dua ruang vektor, yaitu ruang asal (domain) dan ruang tujuan (kodomain). Berikut adalah definisi dan sifat-sifat transformasi linear:

**Definisi Transformasi Linear**:   
- Misalkan $ V $ dan $ W $ adalah ruang vektor. Sebuah pemetaan $ T: V \to W $ disebut transformasi linear jika memenuhi syarat-syarat berikut:
     - $ T(u + v) = T(u) + T(v) $ untuk setiap $ u, v \in V $ (sifat penjumlahan).
     - $ T(ku) = kT(u) $ untuk setiap skalar $ k $ dan $ u \in V $ (sifat perkalian skalar).

## Transformasi Linear Dalam Bentuk Matrix

Matrix $ A $
\begin{equation}
T\left(\left[\begin{array}{l}
2 \\
0
\end{array}\right]\right)=\left[\begin{array}{c}
-2 \\
0
\end{array}\right], \quad T\left(\left[\begin{array}{l}
2 \\
2
\end{array}\right]\right)=\left[\begin{array}{c}
-2 \\
2
\end{array}\right]
\end{equation}


Langkah-langkah untuk menyelesaikan masalah ini adalah :

1. **Misalkan Bentuk Umum Matriks Transformasi \( A \)**:
   \begin{bmatrix} a & b \\ c & d \end{bmatrix}


2. **Gunakan Informasi Transformasi untuk Menentukan \( A \)**:
   Kita tahu bahwa $  T(x) = Ax $ . Maka, kita dapat menuliskan:
   \begin{equation}
T\left(\left[\begin{array}{l}
2 \\
0
\end{array}\right]\right)=A\left[\begin{array}{l}
2 \\
0
\end{array}\right]=\left[\begin{array}{ll}
a & b \\
c & d
\end{array}\right]\left[\begin{array}{l}
2 \\
0
\end{array}\right]=\left[\begin{array}{l}
2 a \\
2 c
\end{array}\right]
\end{equation}

   Dari sini, kita mendapatkan persamaan:

   \begin{bmatrix} 2a \\ 2c \end{bmatrix} = \begin{bmatrix} -2 \\ 0 \end{bmatrix}

   
3. **Gunakan Informasi Transformasi Kedua**:

\begin{equation}
T\left(\left[\begin{array}{l}
2 \\
2
\end{array}\right]\right)=A\left[\begin{array}{l}
2 \\
2
\end{array}\right]=\left[\begin{array}{ll}
a & b \\
c & d
\end{array}\right]\left[\begin{array}{l}
2 \\
2
\end{array}\right]=\left[\begin{array}{l}
2 a+2 b \\
2 c+2 d
\end{array}\right]
\end{equation}

   Dari sini, kita mendapatkan persamaan:

   \begin{bmatrix} 2a + 2b \\ 2c + 2d \end{bmatrix} = \begin{bmatrix} -2 \\ 2 \end{bmatrix}

   

4. **Tentukan Matriks Transformasi \( A \)**: $$ \\ $$
   Dengan $ a = -1 $, $ b = 0 $, $ c = 0 $, dan $ d = 1 $, kita mendapatkan:

   A = 
   \begin{bmatrix} -1 & 0 \\ 0 & 1 \end{bmatrix}


Jadi, matriks transformasi \( A \) yang memenuhi syarat-syarat yang diberikan adalah:

A = 
\begin{bmatrix} -1 & 0 \\ 0 & 1 \end{bmatrix}


## Refleksi Transformasi Linear

Refleksi pada transformasi linear adalah salah satu jenis transformasi yang memindahkan setiap titik pada suatu bidang dengan menggunakan sifat bayangan cermin dari titik-titik yang dipindahkan. $ \\ $


### Rumus Transformasi Refleksi 


1. **Pencerminan terhadap sumbu X**:
   - Matriks transformasi: $$M_X = \begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix}$$
   - Hasil refleksi: $$\mathbf{v'} = M_X \mathbf{v} = \begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} x \\ -y \end{bmatrix}$$

2. **Pencerminan terhadap sumbu Y**:
   - Matriks transformasi: $$M_Y = \begin{bmatrix} -1 & 0 \\ 0 & 1 \end{bmatrix}$$
   - Hasil refleksi: $$\mathbf{v'} = M_Y \mathbf{v} = \begin{bmatrix} -1 & 0 \\ 0 & 1 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} -x \\ y \end{bmatrix}$$

3. **Pencerminan terhadap garis y = x**:
   - Matriks transformasi: $$M_{y=x} = \begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix}$$
   - Hasil refleksi: $$\mathbf{v'} = M_{y=x} \mathbf{v} = \begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} y \\ x \end{bmatrix}$$

4. **Pencerminan terhadap garis y = -x**:
   - Matriks transformasi: $$M_{y=-x} = \begin{bmatrix} 0 & -1 \\ -1 & 0 \end{bmatrix}$$
   - Hasil refleksi: $$\mathbf{v'} = M_{y=-x} \mathbf{v} = \begin{bmatrix} 0 & -1 \\ -1 & 0 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} -y \\ -x \end{bmatrix}$$

5. **Pencerminan terhadap titik asal (pusat koordinat)**:
   - Matriks transformasi: $$M_O = \begin{bmatrix} -1 & 0 \\ 0 & -1 \end{bmatrix}$$
   - Hasil refleksi: $$\mathbf{v'} = M_O \mathbf{v} = \begin{bmatrix} -1 & 0 \\ 0 & -1 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} -x \\ -y \end{bmatrix}$$
